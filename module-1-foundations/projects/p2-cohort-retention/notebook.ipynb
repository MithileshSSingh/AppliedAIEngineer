## Problem Statement
Analyze customer retention by cohort (first purchase month) to understand when and why customers stop purchasing.

```python
import sys
sys.path.insert(0, '../../..')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from shared.viz_utils import setup_style, save_fig
setup_style()
```

## 1. Generate / Load Data

```python
# Synthetic e-commerce transaction data with realistic retention patterns
np.random.seed(42)

n_customers = 500
customers = []

for cid in range(1, n_customers + 1):
    # Each customer's first purchase is in a random month of 2023
    first_month = np.random.randint(1, 13)
    first_date = pd.Timestamp(f'2023-{first_month:02d}-{np.random.randint(1,29):02d}')
    
    # Simulate purchases with decaying probability each subsequent month
    retention_rate = np.random.uniform(0.3, 0.8)  # Varies per customer
    current_date = first_date
    
    for month_offset in range(12):
        purchase_date = first_date + pd.DateOffset(months=month_offset)
        if purchase_date > pd.Timestamp('2024-06-30'):
            break
            
        # First month always has a purchase; subsequent months decay
        if month_offset == 0 or np.random.random() < retention_rate ** (month_offset * 0.5):
            n_orders = np.random.randint(1, 4)  # 1-3 orders that month
            for _ in range(n_orders):
                day = np.random.randint(1, 29)
                date = pd.Timestamp(f'{purchase_date.year}-{purchase_date.month:02d}-{day:02d}')
                revenue = np.random.lognormal(3.5, 1.0)
                customers.append({
                    'customer_id': f'C{cid:04d}',
                    'date': date,
                    'revenue': round(revenue, 2)
                })

df = pd.DataFrame(customers)
df['date'] = pd.to_datetime(df['date'])
print(f"Transactions: {len(df)}, Unique customers: {df['customer_id'].nunique()}")
print(f"Date range: {df['date'].min()} to {df['date'].max()}")
df.head()
```
