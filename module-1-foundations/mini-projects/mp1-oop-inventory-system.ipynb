## Objective
Build an inventory management system using OOP principles learned in the concept notebook. This reinforces: classes, inheritance, dunder methods, dataclasses, and properties.

## Requirements
Build the following classes:

### 1. `Product` (dataclass)
- Fields: `sku` (str), `name` (str), `price` (float), `category` (str)
- Property: `price_with_tax` (8% tax)
- Dunder: `__repr__`, `__eq__` (by SKU), `__lt__` (by price)

### 2. `InventoryItem`
- Wraps a Product with quantity tracking
- Methods: `restock(qty)`, `sell(qty)` (raise error if insufficient)
- Properties: `total_value`, `is_low_stock` (< 10 units)

### 3. `Warehouse`
- Manages a collection of InventoryItems
- Methods: `add_product(product, qty)`, `remove_product(sku)`, `sell(sku, qty)`, `restock(sku, qty)`
- Properties: `total_value`, `low_stock_items`
- Methods: `search(keyword)`, `report()` — print a formatted inventory report
- Dunder: `__len__`, `__contains__` (check by SKU), `__getitem__` (get by SKU)

### 4. `PremiumWarehouse(Warehouse)` (inheritance)
- Applies a 10% premium markup on all prices
- Override `add_product` to apply markup
