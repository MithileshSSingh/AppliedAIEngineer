```python
import sys
sys.path.insert(0, '../..')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler, OneHotEncoder, OrdinalEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer

from shared.viz_utils import setup_style
setup_style()
np.random.seed(42)
```

```python
# Generate a customer churn dataset
n = 2000
df = pd.DataFrame({
    'age': np.random.randint(18, 75, n),
    'income': np.random.lognormal(10.5, 0.8, n).astype(int),
    'credit_score': np.random.normal(650, 80, n).astype(int).clip(300, 850),
    'account_balance': np.random.lognormal(8, 1.5, n).round(2),
    'years_employed': np.random.randint(0, 35, n),
    'num_products': np.random.randint(1, 8, n),
    'has_credit_card': np.random.choice([0, 1], n, p=[0.3, 0.7]),
    'is_active_member': np.random.choice([0, 1], n, p=[0.5, 0.5]),
    'education': np.random.choice(['High School', 'Bachelor', 'Master', 'PhD'], n,
                                   p=[0.30, 0.40, 0.20, 0.10]),
    'employment_type': np.random.choice(['Full-time', 'Part-time', 'Self-employed', 'Unemployed'], n,
                                         p=[0.55, 0.15, 0.20, 0.10]),
    'region': np.random.choice(['North', 'South', 'East', 'West'], n),
    'signup_month': np.random.randint(1, 13, n),
})

# Create a target with known relationships
churn_prob = (
    0.3
    - 0.003 * df['num_products']
    - 0.0003 * df['credit_score']
    + 0.15 * (df['is_active_member'] == 0).astype(float)
    + 0.10 * (df['employment_type'] == 'Unemployed').astype(float)
    - 0.002 * np.log1p(df['account_balance']) / 10
).clip(0.05, 0.85)
df['churned'] = (np.random.random(n) < churn_prob).astype(int)

y = df['churned']
print(f"Shape: {df.shape}, Churn rate: {y.mean():.1%}")
```

```python
# Results tracker
results = []

def evaluate(name, X, y, model=None):
    """Evaluate features with cross-validation and record results."""
    if model is None:
        model = LogisticRegression(max_iter=1000)
    scores = cross_val_score(model, X, y, cv=5, scoring='roc_auc')
    result = {'experiment': name, 'features': X.shape[1], 
              'mean_auc': scores.mean(), 'std_auc': scores.std()}
    results.append(result)
    print(f"{name:40s} | {X.shape[1]:3d} features | AUC: {scores.mean():.4f} (+/- {scores.std():.4f})")
    return scores
```
