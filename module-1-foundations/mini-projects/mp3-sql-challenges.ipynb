```python
import sqlite3
import pandas as pd
import numpy as np

def sql(query, conn):
    return pd.read_sql(query, conn)

def execute(query, conn):
    conn.executescript(query)
    conn.commit()
```

```python
# === DATABASE SETUP (same schema as concept notebook) ===
conn = sqlite3.connect(":memory:")
np.random.seed(42)

execute("""
CREATE TABLE customers (
    customer_id INTEGER PRIMARY KEY, name TEXT, email TEXT,
    tier TEXT, signup_date DATE, region TEXT
);
CREATE TABLE products (
    product_id INTEGER PRIMARY KEY, name TEXT, category TEXT,
    price REAL, cost REAL
);
CREATE TABLE orders (
    order_id INTEGER PRIMARY KEY, customer_id INTEGER,
    order_date DATE, status TEXT
);
CREATE TABLE order_items (
    item_id INTEGER PRIMARY KEY AUTOINCREMENT, order_id INTEGER,
    product_id INTEGER, quantity INTEGER, unit_price REAL, discount REAL DEFAULT 0
);
""", conn)

# Populate (same logic as concept notebook)
names = ['Alice','Bob','Carol','David','Eve','Frank','Grace','Hank','Iris','Jack',
         'Karen','Leo','Mona','Nick','Olivia','Paul','Quinn','Rita','Sam','Tina',
         'Uma','Vic','Wendy','Xander','Yuki','Zara','Amit','Beth','Chris','Dana']
tiers = ['Gold','Silver','Bronze']
regions = ['North','South','East','West']

for i, name in enumerate(names, 1):
    conn.execute("INSERT INTO customers VALUES (?,?,?,?,?,?)",
        (i, name, f"{name.lower()}@email.com", np.random.choice(tiers),
         f"2023-{np.random.randint(1,13):02d}-{np.random.randint(1,29):02d}",
         np.random.choice(regions)))

prods = [(1,'Laptop Pro','Electronics',1299.99,780),(2,'Wireless Mouse','Peripherals',29.99,10),
         (3,'Mech Keyboard','Peripherals',149.99,55),(4,'4K Monitor','Electronics',449.99,220),
         (5,'USB-C Hub','Accessories',59.99,18),(6,'Webcam HD','Peripherals',89.99,30),
         (7,'Standing Desk','Furniture',399.99,180),(8,'Ergo Chair','Furniture',299.99,130),
         (9,'Headset Pro','Accessories',79.99,25),(10,'Laptop Stand','Accessories',49.99,15)]
conn.executemany("INSERT INTO products VALUES (?,?,?,?,?)", prods)

statuses = ['completed']*4 + ['pending','cancelled','returned']
oid = 1
for m in range(1, 13):
    for _ in range(np.random.randint(15, 35)):
        conn.execute("INSERT INTO orders VALUES (?,?,?,?)",
            (oid, np.random.randint(1,31), f"2024-{m:02d}-{np.random.randint(1,29):02d}",
             np.random.choice(statuses)))
        for pid in np.random.choice(range(1,11), np.random.randint(1,4), replace=False):
            conn.execute("INSERT INTO order_items (order_id,product_id,quantity,unit_price,discount) VALUES (?,?,?,?,?)",
                (oid, int(pid), np.random.randint(1,5), prods[pid-1][3],
                 np.random.choice([0,0,0,0.05,0.10,0.15])))
        oid += 1
conn.commit()
print("Database ready!")
for t in ['customers','products','orders','order_items']:
    print(f"  {t}: {conn.execute(f'SELECT COUNT(*) FROM {t}').fetchone()[0]} rows")
```
